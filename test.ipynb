```python
import numpy as np
```

```python
kernel = np.array([
    [1, 2, 1],
    [2, 4, 2],
    [1, 2, 1]
])
```

```python
sigma = 1.3
ksize = 3
center = ksize // 2

new_kernel = np.zeros(shape=(ksize, ksize), dtype=np.float32)

for x in range(0, ksize):
    x2 = (x - center) ** 2
    for y in range(0, ksize):
        y2 = (y - center) ** 2
        g = np.exp(-(x2 + y2) / (2 * sigma * sigma))
        g = g / (2 * np.pi * sigma * sigma)
        new_kernel[x, y] = g
```

```python
new_kernel / new_kernel.sum()
```
